```python
csv_file = "./Basic_Stats.csv"
basic_stats_df = pd.read_csv(csv_file)
basic_stats_df.head()
```

```python
basic_stats_df = basic_stats_df[['age'
                                ,'player_id'
                                ,'position'
                                ,'name'
                                ,'current_status'
                                ,'experience'
                                ,'current_team']].copy()
basic_stats_df
```

```python
WR_and_TE_df = pd.read_csv("./Game_Logs_Wide_Receiver_and_Tight_End.csv")
WR_and_TE_df
```

```python
new_WR_and_TE_df = WR_and_TE_df[['player_id'
                                 ,'name'
                                 ,'position'
                                 ,'year'
                                 ,'outcome'
                                 ,'opponent'
                                 ,'home_or_away'
                                 ,'score'
                                 ,'receptions'
                                 ,'receiving_yards'
                                 , 'yards_per_reception'
                                 ,'longest_reception'
                                 , 'receiving_tds'
                                 , 'rushing_attempts'
                                 , 'rushing_yards'
                                 , 'yards_per_carry'
                                 , 'longest_rushing_run'
                                 , 'rushing_tds'
                                 , 'fumbles'
                                 , 'fumbles_lost']].copy()
new_WR_and_TE_df
```

Create TABLE nfl_basic_stats(
	age int,
	player_id varchar primary key,
	position varchar,
	name varchar,
	current_status varchar,
	experience varchar,
	current_team varchar
	);
	
Create Table nfl_game_logs_WR_and_TE(
	Player_Id varchar,
	Name varchar,
	Position varchar,
	Year int,
	Outcome varchar,
	Opponent varchar,
	Home_or_Away varchar,
	Score varchar,
	Receptions varchar,
	Receiving_Yards varchar,
	Yards_Per_Reception varchar,
	Longest_Reception varchar,
	Receiving_TDs varchar,
	Rushing_Attempts varchar,
	Rushing_yards varchar,
	Yards_Per_Carry varchar,
	Longest_Rushing_Run varchar,
	Rushing_TDs varchar,
	Fumbles varchar,
	Fumbles_Lost varchar);

select * from nfl_basic_stats
select * from nfl_game_logs_wr_and_te

```python
connection_string = f'postgres:{password}@localhost:5432/NFL_stats'
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
basic_stats_df.to_sql(name = "nfl_basic_stats", con = engine, if_exists = 'append', index = False)
```

```python
new_WR_and_TE_df.to_sql(name = "nfl_game_logs_wr_and_te", con = engine, if_exists = 'append', index = False)
```

```python
pd.read_sql_query('select * from nfl_basic_stats', con=engine)
```

```python
pd.read_sql_query('select * from nfl_game_logs_wr_and_te', con=engine)
```
